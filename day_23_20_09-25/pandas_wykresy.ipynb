```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

bins = range(0, int(df['Calories'].max()) + 100, 100)
df['Calories_bin'] = pd.cut(df['Calories'], bins=bins, right=False)

g = sns.catplot(
    data=df,
    x="Calories_bin", y="Pulse",
    kind="box",
    height=5,
    aspect=2
)

plt.xticks(rotation=60)  # obrt etykiet dla lepszej czytelnoci
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# binowanie po PULSIE zamiast Kalorii
bins = range(0, int(df['Pulse'].max()) + 10, 10)  # co 10 uderze
df['Pulse_bin'] = pd.cut(df['Pulse'], bins=bins, right=False)

g = sns.catplot(
    data=df,
    x="Pulse_bin",    # kategorie = przedziay pulsu
    y="Calories",     # wartoci liczbowe
    kind="box",
    height=5,
    aspect=2
)

plt.xticks(rotation=60)
plt.show()
```

```python
bins = range(0, int(df['Calories'].max()) + 100, 100)
df['Calories_bin'] = pd.cut(df['Calories'], bins=bins, right=False)

g = sns.catplot(
    data=df,
    x="Calories_bin", y="Pulse",
    kind="violin",
    height=5,
    aspect=2
)
```
